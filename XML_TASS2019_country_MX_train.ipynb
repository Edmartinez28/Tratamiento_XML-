```python
import xml.etree.ElementTree as et
import pandas as pd 

xdata = et.parse(r"C:\Users\ASUS\Documents\UTPL\cuarto ciclo\Practicum 1.2\Semana 9\Info\Info\TASS2019_country_MX_train.xml")
xroot = xdata.getroot()

df_colums = ["tweet","tweetid","user","content","date","lang","sentiment"] 
rows = []

for node in xroot: 
    s_tweet = node.attrib.get("tweet")
    s_tweetid = node.find("tweetid").text if node is not None else None
    s_user = node.find("user").text if node is not None else None
    s_content = node.find("content").text if node is not None else None
    s_date = node.find("date").text if node is not None else None
    s_lang = node.find("lang").text if node is not None else None
    s_sentiment = node.find("sentiment").find("polarity").find("value").text if node is not None else None
    
    rows.append({"tweet": s_tweet, "tweetid": s_tweetid, 
                 "user": s_user, "content": s_content,
                 "date": s_date, "lang": s_lang
                ,"sentiment":s_sentiment})
    
out_df = pd.DataFrame(rows, columns = df_colums)
```
